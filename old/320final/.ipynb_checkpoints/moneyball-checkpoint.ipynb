<h4>Part one</h4>
Wrote 2 queries - one to sum up the wins and losses of each team for each year, and one to sum up the salaries for 
each team and year. Then I joined the two tables together by matching teamID and year. To handle missing data from either table, I used inner join to join the two which only includes full matches, so if any table had missing data the row was not included in the final pandas data frame.


```python
#SQL QUERY
q = """
    SELECT
        sals.yearID,
        sals.teamID,
        total_wins,
        total_losses,
        total_payroll,
        CAST(total_wins as FLOAT)/(total_wins+total_losses)*100 as winp
        FROM
        (
            SELECT yearID, teamID, sum(w) as total_wins, sum(l) as total_losses
            FROM
                Teams
            GROUP BY teamID, yearID) teams
            JOIN
                (
                    SELECT
                        yearID,teamID, sum(salary) as total_payroll
                    FROM Salaries
                    GROUP BY teamID,yearID) sals
                    ON teams.teamID = sals.teamID AND teams.yearID = sals.yearID
"""


teams_and_sals = pandas.read_sql(q, conn)
```

<p>Now let's standardize the payrolls</p>

```python
#create dataframe with avg and std of payrolls for each year.
avg_payrolls_per_year = teams_and_sals[['yearID', 'total_payroll']]
avg_payrolls_per_year = avg_payrolls_per_year[avg_payrolls_per_year.yearID > 1989] 

#create two data frames - one with standard dev and one with mean for each year
STDdf = avg_payrolls_per_year.groupby(['yearID']).std()
STDdf.rename(columns={'total_payroll': 'std'}, inplace=True)
MEANdf = avg_payrolls_per_year.groupby(['yearID']).mean()
MEANdf.rename(columns={'total_payroll': 'mean'}, inplace=True)

#merge the two into one
mean_std = pandas.concat([STDdf,MEANdf], axis=1)
mean_std.reset_index(inplace=True)
#print for the world to see
mean_std
```

```python
#seperate into bins - same as above
bins = [1990,1995,2000,2005,2010,2014]
labels = ['1990-1995', '1995-2000', '2000-2005', '2005-2010', '2010-2014']
graph = teams_and_sals[['yearID', 'teamID', 'total_payroll','winp']]
graph = graph[graph.yearID > 1989] 

#slice (cut) and dice baby
graph['periodBinned'] = pandas.cut(graph['yearID'], bins, include_lowest=True)

```

```python
#merge the mean_std datable withthe binned table from above
graph = pandas.merge(graph,mean_std, on='yearID')
graph
#graph with std and mean for each team/year combo
```

```python
#calc and set standard payroll for each year/team combo
for i, row in graph.iterrows():
    sp = (row[2]-row[6])/row[5]
    graph.set_value(i,'standardized_payroll',sp)
```

<p>Now I'm going to add the expected win percentage and the efficiency for each team into the dataframe</p>

```python
graph.reset_index(inplace=True)
#Adding expected win percentage to dataframe
for i, row in graph.iterrows():
    expected_winp = 50 + (2.5 * row[4])
    graph.set_value(i,'expected_winp',expected_winp)
#Adding efficiency to dataframe
for i, row in graph.iterrows():
    efficiency = row[3] - row[5]
    graph.set_value(i,'efficiency',efficiency)
```
